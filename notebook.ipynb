```python
import pandas as pd
```

```python
# Load the data
# characters = pd.read_csv('characters.csv')
df = pd.read_csv('episodes.csv', delimiter=';', date_parser=['Airdate'], dayfirst=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of ratings
fig = plt.figure(figsize=(16, 9))
plt.hist(df['Rating'], bins=20)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Extract the season from the airdate
df['Season'] = pd.DatetimeIndex(df['Airdate']).year

# Calculate the average rating by season
season_ratings = df.groupby('Season')['Rating'].mean()

# Plot the average rating by season
fig = plt.figure(figsize=(16, 9))
plt.plot(season_ratings.index, season_ratings.values)
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.title('Average Rating by Season')
plt.show()
```
